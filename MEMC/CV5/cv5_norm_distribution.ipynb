```python
#TOHLE PREDELAT MA BEJT KOULE
# Generovani bodu na sfere:
SAMPLE = 2000
rand_matrix = np.random.rand(2, SAMPLE)
xx = rand_matrix[0] * np.pi
yy = rand_matrix[1] * 2 * np.pi

x = np.sin(xx) + np.cos(yy)
y = np.sin(xx) + np.sin(yy)
z = np.cos(xx)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(x, y, z)
plt.show()
```

```python
SAMPLE = 1000
mu = np.random.uniform(-1, 1, SAMPLE)
phi = np.random.uniform(0, 2*np.pi, SAMPLE)
sqrt = 1-mu**2
x = np.sqrt(sqrt) * np.sin(phi)
y = np.sqrt(sqrt) * np.cos(phi)
z = mu
```

```python

from mpl_toolkits.mplot3d import Axes3D

plt.clf()
plt.figure()
fig = plt.figure(figsize=(6,6))
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)
ax.scatter(x, y, z, s=40, c=x, marker='o', alpha=1)
plt.savefig("scatter_hue", bbox_inches='tight')
```

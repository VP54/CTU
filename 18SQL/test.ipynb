```python
import numpy as np
import pandas as pd
from faker import Faker
import random
from datetime import datetime, timedelta


import random



# Define score ranges for sentiments
scores = {
    'bad': random.randint(1, 3),
    'terrible': random.randint(1, 3),
    'poor': random.randint(2, 4),
    'awful': random.randint(2, 4),
    'good': random.randint(4, 7),
    "okay": random.randint(3, 6),
    'great': random.randint(6, 9),
    'fantastic': random.randint(8, 10),
    'excellent': random.randint(8, 10),
    'awesome': random.randint(9, 10),
    'horrible': random.randint(1, 3)
}

# Initialize Faker to generate realistic data
fake = Faker()

# Function to generate random data for each row
def generate_random_data(scores):
    review_id = random.randint(100000, 999999)
    random_sentiment = random.choice(list(scores.keys()))
    review_text = random_sentiment
    score = scores[random_sentiment]
    start_date = datetime.now() - timedelta(days=365)
    end_date = datetime.now()
    date = fake.date_time_between_dates(start_date, end_date)
    return [review_id, date, score, review_text]

# Generate random data for a single row
data = [generate_random_data(scores) for _ in range(100)]  # Change '10' to the desired number of rows

# Create a Pandas DataFrame
columns = ['review_id', 'date', "score", "review_text"]
df = pd.DataFrame(data, columns=columns)


df.to_csv("reviews.csv")
```
